```python
# Calculate mean for each question response for all the classes.
questionmeans = []
classlist = []
questions = []
totalplotdata = pd.DataFrame(list(zip(classlist,questions,questionmeans))
                      ,columns=['class','questions', 'mean'])
for class_num in range(1,13):
    class_data = df[(df["class"]==class_num)]
    
    questionmeans = []
    classlist = []
    questions = []
    
    for num in range(1,13):
        questions.append(num)
    #Class related questions are from Q1 to Q12
    for col in range(5,17):
        questionmeans.append(class_data.iloc[:,col].mean())
    classlist += 12 * [class_num] 
    print(classlist)
    plotdata = pd.DataFrame(list(zip(classlist,questions,questionmeans))
                      ,columns=['class','questions', 'mean'])
    totalplotdata = totalplotdata.append(plotdata, ignore_index=True)
```

```python
plt.figure(figsize= (14,10))
sns.pointplot(x = 'questions', y = 'mean', data = totalplotdata, hue = 'class')
```

### Lets begin to cluster the students based on the questionaire data

```python
#Lets try to cluster all the students based on the Question responses data
data_questions = df.iloc[:,5:33]
data_questions.head()
```

```python
#lets do a PCA for feature dimensional reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_questions_pca = pca.fit_transform(data_questions)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 7):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data_questions_pca)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize = (14,10))    
plt.plot(range(1, 7), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Based on the Elbow graph , we can go for 3 clusters.m

```python
kmeans = KMeans(n_clusters= 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(data_questions_pca)
```

```python
#Visualising the clusters
plt.figure(figsize = (14,10))
plt.scatter(data_questions_pca[y_kmeans == 0,0], data_questions_pca[y_kmeans ==0,1], s = 100, c = 'y', label = 'Cluster1')

plt.scatter(data_questions_pca[y_kmeans == 1,0], data_questions_pca[y_kmeans ==1,1], s = 100, c = 'g', label = 'Cluster2')

plt.scatter(data_questions_pca[y_kmeans == 2,0], data_questions_pca[y_kmeans ==2,1], s = 100, c = 'r', label = 'Cluster3')
            
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = 'b', label = "Centroids")            
            
plt.title("Clusters of Students")            
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")            
plt.legend()
plt.show()            
```

Looking at the above graph , i see we have 3 clusters of students who have given like Negative, Neutral and Positive feedback

```python
# Let me check the count of students in each cluster
import collections
collections.Counter(y_kmeans)
```

So we have 2358 students who have given negative ratings overall , 2222 students with positive ratings and 1240 students with nuetral response

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage = 'ward')
y_hc = hc.fit_predict(data_questions_pca)
X = data_questions_pca

#Visualising the Clusters
plt.figure(figsize= (14,10))
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s = 100, c = 'y', label = 'Cluster1')

plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s = 100, c = 'r', label = 'Cluster2')

plt.title("Cluster of students")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
plt.show()
```
